```python
import numpy as np
import pandas as pd
import pydotplus
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
```

```python
dataset=pd.read_csv('/content/data kami.csv', delimiter=',', header=0)
dataset.head()
```

```python
x1=dataset.drop('Status', axis=1)
x=x1.drop('Country',axis=1)
y=dataset['Status']
x_train,x_test,y_train,y_test = train_test_split(x,y, train_size=0.7, random_state=0)
```

```python
x_train = x_train.applymap(lambda x: float(str(x).replace(',', '.')) if isinstance(x, str) else x)
```

```python
hasil=KNeighborsClassifier(n_neighbors=3)
hasil_train=hasil.fit(x_train,y_train)
```

```python
x_test = x_test.applymap(lambda x: float(str(x).replace(',', '.')) if isinstance(x, str) else x)
```

```python
y_pred=hasil_train.predict(x_test)
```

```python
print('akurasi:', accuracy_score(y_test,y_pred))
```

```python
y_pred=np.round(hasil_train.predict(x_test))
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
print('Hasil:' , classification_report(y_test,y_pred))
```
